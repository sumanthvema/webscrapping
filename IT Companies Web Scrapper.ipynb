```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[1]

all_titles = table.find_all('th')
```

```python
titles = [title.text.strip() for title in all_titles]

print(titles)
```

```python
import pandas as pd

df = pd.DataFrame(columns = titles)

df
```

```python
column_data = table.find_all('tr')

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_data
```
